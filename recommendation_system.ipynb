```python
import numpy as np
import pandas as pd
# from text_processing import tasting_notes_words_filtered
```

```python
df = pd.read_csv('vectors_and_tasting_notes.csv')
df
```

```python
df['tasting_note'].unique()
```

```python
new_vector_tasting_notes = ['blackberry', 'jam', 'berry', 'raisin', 'cherry', 'sherry']
new_vector = [1 if tasting_note in new_vector_tasting_notes else 0 for tasting_note in df['tasting_note'].unique()]

vectors = df['review_text_vector'].values
vectors = np.vstack(vectors)

distances = np.linalg.norm(vectors - new_vector, axis=1)

nearest_index = np.argmin(distances)
nearest_vector = vectors[nearest_index]
nearest_distance = distances[nearest_index]
```

```python
recommended_bottle_title = df.loc[nearest_index, 'review_title']

df[df['review_title'] == recommended_bottle_title]
```
